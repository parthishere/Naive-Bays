```python
feature_cols = ['continent', 'date_added','release_year', 'rating','duration','genre']
X = netflix_ds[feature_cols] # Features
y = netflix_ds.type # Target variable
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)# Logistic Regression
X_train.shape, y_train.shape, X_test.shape
```

```python
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_train, y_train) * 100, 2)
print(acc_gaussian)
cf_matrix = confusion_matrix(y_test,Y_pred)
sns.heatmap(cf_matrix, annot=True)
```

```python
Stochastic_gradient_descent = SGDClassifier(loss="hinge", penalty="l2", max_iter=5)
Stochastic_gradient_descent.fit(X_train, y_train)
Y_pred = Stochastic_gradient_descent.predict(X_test)
acc_log = round(Stochastic_gradient_descent.score(X_train, y_train) * 100, 2)
print(acc_log)
cf_matrix = confusion_matrix(y_test,Y_pred)
sns.heatmap(cf_matrix, annot=True)
```
